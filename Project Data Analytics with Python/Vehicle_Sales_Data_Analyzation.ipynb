```python
import pandas as pd
evp = pd.read_csv('Electric_Vehicle_Population_Data.csv')
evp.drop_duplicates(subset='VIN (1-10)', inplace=True)
evp
```

***เริ่มการหาคำตอบ ข้อที่ 1 : BEV และ PHEV มีกี่คัน จากข้อมูลชุดนี้***

***จากโจทย์ข้อที่ 1***

สามารถตอบคำถามได้ว่า

จำนวนรถประเภท BEV มีทั้งหมด 5,341 คัน

และรถประเภท PHEV มีทั้งหมด 4,188 คัน

```python
from matplotlib import pyplot as plt
import numpy as np

fig = plt.figure(figsize=(10,5))
```

```python
cars = np.array([5341, 4188])
cars_labels = ["BEV","PHEV"]
type_colors = ["r", "b"]
cars_explode = [0.1, 0]

plt.pie(cars, labels = cars_labels, explode = cars_explode,colors = type_colors, shadow = True)
plt.show()
```

***เริ่มการหาคำตอบข้อที่ 2 : เรียงลำดับยี่ห้อรถประเภท BEV ที่มีการใช้งานมากที่สุด 10 อันดับแรก ใน WA State, USA***

```python
bev_list = evp[evp['Electric Vehicle Type'].apply(lambda member: 'BEV' in member)]
bev_list
```

```python
bevMaker = bev_list['Make']
bevMaker
```

```python
bevMaker.apply(lambda m: m.split(', ')).explode().unique()
```

```python
bevMakerExplode = bevMaker.explode('Make')
bevMakerExplode
```

```python
counts_maker = bevMakerExplode.value_counts().nlargest(10)
counts_maker
```

```python
cars_amount = [1303,524,513,365,336,327,294,266,226,221]
cars_maker = ['TESLA','AUDI','FORD','NISSAN','HYUNDAI','VOLKSWAGEN','CHEVROLET','KIA','BMW','MERCEDES-BENZ']
cars_explode2 = [0.05, 0.05, 0.05, 0.05, 0.05,0.05, 0.05, 0.05, 0.05, 0.05]

plt.pie(cars_amount, labels = cars_maker, explode = cars_explode2, autopct='%1.1f%%', pctdistance=0.85, shadow = True)

centre_circle = plt.Circle((0, 0), 0.40, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.title('Top 10 Car Brands (BEV) in WA State, USA in 2023')
plt.show()
```
